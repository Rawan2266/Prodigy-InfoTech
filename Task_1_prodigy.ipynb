```python
mode_region = df_cou['Region'].mode()[0]
df_cou['Region'] = df_cou['Region'].fillna(mode_region)
print(df_cou['Region'].isnull().sum())
```

```python
mode_income = df_cou['IncomeGroup'].mode()[0]
df_cou['IncomeGroup'] = df_cou['IncomeGroup'].fillna(mode_income)
print(df_cou['IncomeGroup'].isnull().sum())
```

```python
mode_specNote = df_cou['SpecialNotes'].mode()[0]
df_cou['SpecialNotes'] = df_cou['SpecialNotes'].fillna(mode_specNote )
print(df_cou['SpecialNotes'].isnull().sum())
```
